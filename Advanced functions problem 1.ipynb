```python
def innermost(list): #define the function name
    iteration = list #create the iteration for the while loop so we can exit loop once done

    while isinstance(iteration[-1], list): #create while loop, using the list and the iteration to stop the loop once done
        iteration = iteration[-1] #this will remove 1 tick from the notations until the loop stops

    #
    return [x + 1 for x in iteration] #completes the loop


list = [1, 2, 3, 4, [5, 6, 7, [888, 999]]] #creates the list
print(innermost(list))  #call the function and print result
```
