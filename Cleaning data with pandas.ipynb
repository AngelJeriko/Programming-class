```python
# -*- coding: utf-8 -*-

#%% Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#%% Importing Data
flights = pd.read_csv(r"C:\Users\kanak\Downloads\flights.csv")
flights.head(10)
weather_data_pd = pd.read_csv(r'C:\Users\kanak\Downloads\weather.csv')
weather_data_np = weather_data_pd.to_numpy()
#%% Pandas Data Filtering/Sorting Question Answering
#use flights_data

def have(df, *cols):
    missing = [c for c in cols if c not in df.columns]
    if missing:
        raise KeyError(f"Missing columns: {missing}\nAvailable: {df.columns.tolist()}")

# #Question 1 How many flights were there from JFK to SLC? Int
have(flights, "origin", "dest")
q_1 = len(flights[(flights["origin"] == "JFK") & (flights["dest"] == "SLC")])
print("Answer for q_1:", q_1)

# #Question 2 How many airlines fly to SLC? Should be int

have(flights, "dest", "carrier")
q_2 = flights.loc[flights["dest"] == "SLC", "carrier"].nunique()
print("Answer for q_2:",q_2)

# #Question 3 What is the average arrival delay for flights to RDU? float

have(flights, "dest", "arr_delay")
q_3 = flights.loc[flights["dest"] == "RDU", "arr_delay"].mean()
print("Answer for q_3:",q_3)

# #Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
have(flights, "dest", "origin")
to_sea = flights[flights["dest"] == "SEA"]
nyc_to_sea = to_sea[to_sea["origin"].isin(["JFK", "LGA"])]
q_4 = len(nyc_to_sea) / len(to_sea)
print("Answer for q_4:",q_4)
# #Question 5 Which date has the largest average depature delay? Pd slice with date and float
# #please make date a column. Preferred format is 2013/1/1 (y/m/d)
have(flights, "year", "month", "day", "dep_delay")
flights["date"] = pd.to_datetime(flights[["year", "month", "day"]])
avg_dep_delay = flights.groupby("date")["dep_delay"].mean()
q_5 = avg_dep_delay.sort_values(ascending=False).head(1)
q_5.index = q_5.index.strftime("%Y/%m/%d")  # format like 2013/1/1
print("Answer for q_5:",q_5)

# #Question 6 Which date has the largest average arrival delay? pd slice with date and float
have(flights, "year", "month", "day", "arr_delay")
avg_arr_delay = flights.groupby(pd.to_datetime(flights[["year", "month", "day"]]))["arr_delay"].mean()
q_6 = avg_arr_delay.sort_values(ascending=False).head(1)
q_6.index = q_6.index.strftime("%Y/%m/%d")
print("Answer for q_6:",q_6)
# #Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
# #speed = distance/airtime
have(flights, "origin", "year", "air_time", "distance", "tailnum")
# Calculate speed in miles per hour
flights["speed"] = flights["distance"] / (flights["air_time"] / 60)
flights_2013 = flights[(flights["year"] == 2013) & (flights["origin"].isin(["LGA", "JFK"]))]
q_7 = flights_2013[["tailnum", "speed"]].sort_values("speed", ascending=False).head(1)
print("Answer for q_7:",q_7)

# #Question 8 Replace all nans in the weather pd dataframe with 0s. Pd with no nans
# Replace NaNs in weather data with 0
q_8 = weather_data_pd.fillna(0)
q_8.isna().sum().sum()  # should print 0

# #%% Numpy Data Filtering/Sorting Question Answering
# #Use weather_data_np
# #Question 9 How many observations were made in Feburary? Int
# How many observations were made in February?
# Identify column indexes
month_idx = np.where(weather_data_pd.columns.str.lower() == "month")[0][0]
is_feb = weather_data_np[:, month_idx] == 2
q_9 = np.sum(is_feb)
print("Answer for q_9:",q_9)
# #Question 10 What was the mean for humidity in February? Float
# Mean humidity in February
humid_idx = np.where(weather_data_pd.columns.str.lower().isin(["humid","humidity"]))[0][0]
feb_humid = weather_data_np[is_feb, humid_idx].astype(float)
q_10 = np.nanmean(feb_humid)
print("Answer for q_10:",q_10)

# #Question 11 What was the std for humidity in February? Float
# Std of humidity in February
q_11 = np.nanstd(feb_humid)
print("Answer for q_11:",q_11)
```
