```python
#question 1
import numpy as np
print(np.__version__)

#question 2
arr = np.arange(10)
print(arr)

# pull the data set from link provided
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

iris = np.genfromtxt(url, delimiter=',', dtype=None, encoding='utf-8')

# quesiton 3 print first seven rows to show it was correctly pulled
print(iris[:7])

petal_width = iris['f3']

# question 4 Find the position of the first occurrence where value > 1.0 in the fourth column of the provided dataset
position = np.where(petal_width > 1.0)[0][0]

print("Position of first value > 1.0 in petal width column:", position)
print("Value at that position:", petal_width[position])
print("Species at that position:", iris['f4'][position])

#question 5
# Generate the random array
np.random.seed(100)
a = np.random.uniform(1, 50, 20)
print("Original array:\n", a)

# Replace values using np.clip()
a_clipped = np.clip(a, 10, 30)

print("\nModified array:\n", a_clipped)
```
